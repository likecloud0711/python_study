```python
import numpy as np

t = np.arange(0, 12, 0.01)
print(t)
y = np.sin(t)
print(y)
```

```python
plt.figure(figsize=(6,4))
plt.plot(t, y)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(t, y)
plt.grid() #  
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(t, y)
plt.grid() #  
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(t, y, label='sin')
plt.plot(t, np.cos(t), color='r', label='cos') #  
plt.legend() # label
plt.grid() #  
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.show()
```

```python
t = [0, 1, 2, 3, 4, 5, 6]
y = [1, 2, 5, 8, 9, 5, 3]
fig = plt.figure(figsize=(10,6))
ax1 =fig.add_subplot(2,2,1)
ax1.plot(t, y, color='g')
ax2 =fig.add_subplot(2,2,2)
ax2.plot(t, y, color='b', linestyle='--')
ax3 =fig.add_subplot(2,2,3)
ax3.plot(t, y, color='k', linestyle='--', marker='o')
ax4 =fig.add_subplot(2,2,4)
ax4.plot(t, y, color='r', linestyle='--', marker='o', markerfacecolor='b')

plt.plot()
```

```python
s1 = np.random.normal(loc=0, scale=1, size = 1000)
s2 = np.random.normal(loc=5, scale=0.1, size = 1000)
s3 = np.random.normal(loc=10, scale=2, size = 1000)
```

```python
plt.figure(figsize=(10,6))
plt.plot(s1, label ='s1')
plt.plot(s2, label ='s2')
plt.plot(s3, label ='s3')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.boxplot((s1,s2,s3))
plt.grid()
plt.show()
```
