```python
# 결손 데이터
data = {'name':['kim','lee','choi','park'],
       'score' : [100,95,90,85],
       'grade' : ['A','A','B','B']
       }
df = pd.DataFrame(data)
df
```

```python
df['point'] = np.nan
df
```

```python
#NaN값 체크
df.isnull() # isna()
```

```python
df.isna()
```

```python
# NaN값을 0으로 채우기
df.fillna(value = 0, inplace=True)
df
```

```python
df['point2'] = [50,100,70,np.nan]
df
```

```python
# 평균값으로 채우기
df.fillna(value=df['point2'].mean())
```

```python
# NaN 값 제거
# df.dropna(how='any')
# df.dropna(how='all')
df.dropna(how='all')
```

```python
# 행렬 바꾸기
df = df.T
df
```

```python
d1 = [{'name':'john', 'job':'teacher'},
         {'name':'Nate', 'job':'student'},
         {'name':'Fred', 'job':'developer'}
     ]
d2 = [{'name':'ED', 'job':'dentist'},
         {'name':'Jack', 'job':'farmer'},
         {'name':'Ted', 'job':'designer'}
     ]

df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2, columns=['name','job'])
df1
```

```python
df2
```

```python
frames = [df1,df2]
result = pd.concat(frames, ignore_index=True)
result
```

```python
result = pd.concat([df1,df2], axis=1, ignore_index=True)
result
```

```python
m_df = pd.DataFrame({'고객번호':[1001,1002,1003,1004,1005],
                    '이름':['AAA','BBB','CCC','DDD','EEE']}
                    ,columns=['고객번호','이름'])
m_df
```

```python
b_df = pd.DataFrame({'고객번호':[1001,1001,1003,1004,1006],
                    '금액':[1000,2000,1500,500,700]}
                    ,columns=['고객번호','금액'])
b_df
```

```python
inner_result = pd.merge(m_df,b_df,how='inner')
inner_result
```

```python
inner_result = pd.merge(m_df,b_df,how='outer')
inner_result
```

```python
inner_result = pd.merge(m_df,b_df,how='right')
inner_result
```

```python
inner_result = pd.merge(m_df,b_df,how='left')
inner_result
```
